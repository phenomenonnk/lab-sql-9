## Create a table rentals_may to store the data from rental table with information for the month of May.

```python
# Create a Python connection with SQL database and retrieve
# connect python with an sql database and getting the results of the queries as dataframes
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

data = pd.read_sql_query('SELECT * FROM rental', engine)
data.head()
```

## Check the number of rentals for each customer for May.

```python
rentals_may_count_for_each_customer_id=rentals_may.groupby(['customer_id'])['customer_id'].count()
print(rentals_may_count_for_each_customer_id)
```

```python
# or
rentals_may_count_for_each_customer_id_query = pd.read_sql_query('''select customer_id, count(customer_id) as number_of_rentals_for_each_customer_id_may from rental
where substring(rental_date, 6, 2)= 05
group by customer_id;''', engine)
rentals_may_count_for_each_customer_id_query 
```

## Check the number of rentals for each customer for June.

```python
rentals_june_count_for_each_customer_id=rentals_june.groupby(['customer_id'])['customer_id'].count()
print(rentals_june_count_for_each_customer_id)
```

```python
rentals_june_count_for_each_customer_id_query = pd.read_sql_query('''select customer_id, count(customer_id) as number_of_rentals_for_each_customer_id_may from rental
where substring(rental_date, 6, 2)= 06
group by customer_id''', engine)
print(rentals_june_count_for_each_customer_id_query) 
```
